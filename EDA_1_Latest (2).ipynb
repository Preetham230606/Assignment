```python
#Load the libraries
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data =pd.read_csv("data_clean.csv")
```

```python
data
```

```python
data.info()
```

```python
print(type(data))
print(data.shape)
print(data.size)
```

```python
# Drop Duplicate column (Temp c) and Unamed column

# data1 = data.drop(['Unnamed: 0', "Temp C"], axis =1, inplace = True)
data1 = data.drop(['Unnamed: 0',"Temp C"], axis =1)
data1
```

```python
data1.info()
```

```python
#Convert the Month column  data type to float data type

data1['Month']=pd.to_numeric(data['Month'],errors='coerce')
data1.info()
```

```python
#Checking for duplicated rows in the table
#Print only the duplicated row (one) only
data1[data1.duplicated(keep = False)]
```

```python
#Drop duplicated rows
data1.drop_duplicates(keep='first', inplace = True)
data1
```

```python
data1.rename({'Solar.R':'Solar'}, axis=1, inplace = True)
data1
```

```python
data1.isnull().sum()
```

```python
#Visualize data1 missing values heat map

cols = data1.columns
colors = ['white', 'blue']
sns.heatmap(data1[cols].isnull(),cmap=sns.color_palette(colors),cbar =True)
```

```python
#Find the mean and media values of each numeric column
#Imputation of missing value with median
median_ozone = data1["Ozone"].median()
mean_ozone = data1["Ozone"].mean
print("Median of Ozone: ", median_ozone)
print("Mean of Ozone: ", mean_ozone)
```

```python
#Replace the Ozone missing values with median values
data1['Ozone'] = data1['Ozone'].fillna(median_ozone)
data1.isnull().sum()
```

```python
#Find the mean and media values of each numeric column
#Imputation of missing value with median
median_Solar = data1["Solar"].median()
mean_Solar = data1["Solar"].mean()
print("Median of Solar: ", median_Solar)
print("Mean of Ozone: ", mean_Solar)
```

```python
#Replace the Ozone missing values with median values
data1['Solar'] = data1['Solar'].fillna(median_Solar)
data1.isnull().sum()
```

```python
data1.head()
```

```python
#Find the mde values of categorical columns(weather)

print(data1["Weather"].value_counts())
mode_weather = data1["Weather"].mode()[0]
print(mode_weather)
```

```python
data1["Weather"] = data1["Weather"].fillna(mode_weather)
data1.isnull().sum()
```

```python
mode_month=data1["Month"].mode()[0]
data1["Month"]=data1["Month"].fillna(mode_month)
data1.isnull().sum()
```

```python
data1.tail()
```

```python
#Reset the index column
data1.reset_index(drop=True)
```

##### Detecetion of outliers in the columns

##### Method 1: Using Histogram boxplot

```python
fig,axes=plt.subplots(2,1,figsize=(8,6),gridspec_kw={'height_ratios':[1,3]})
sns.boxplot(data=data1["Ozone"],ax=axes[0],color='skyblue',width=0.5,orient='h')
axes[0].set_title("Boxplot")
axes[0].set_xlabel("Ozone Levels")

sns.histplot(data1["Ozone"],kde=True,ax=axes[1],color='Purple',bins=30)
axes[1].set_title("Histogram with KDE")
axes[1].set_xlabel("Ozone Levels")
axes[1].set_ylabel("Frequency")
plt.tight_layout()
plt.show()
```

##### Observations
- The ozone column has extreme values beyond 81 as seen from box plot
- The same is confirmed from the below right=skewed histogram

```python
# Create a figure for violin plot
sns.violinplot(data=data1["Ozone"].color='lightgreen')  
plt.title("Violin plot")
plt.show
```
